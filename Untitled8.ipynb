```python
import pandas as pd
```

```python
path = "/content/Relatorio 302 (Tratado).xlsx"
path2 = '/content/CA - Contas a Receber - MAG - 31102024.xls'

df = pd.read_excel(path)
df2 = pd.read_excel(path2)

df = df.drop(columns='Validade')

def formatar_cnpj(cnpj):
    cnpj = str(cnpj).zfill(14)  # Garante que tenha 14 dígitos
    return f"{cnpj[:2]}.{cnpj[2:5]}.{cnpj[5:8]}/{cnpj[8:12]}-{cnpj[12:]}"

# Aplicando a formatação à coluna
df2['Identificador do cliente'] = df2['Identificador do cliente'].apply(formatar_cnpj)
# Renomeando a coluna 'Identificador do cliente' para 'CNPJ'
df2.rename(columns={'Identificador do cliente': 'CNPJ'}, inplace=True)
df2.rename(columns={'Nome do cliente': 'Razão Social'}, inplace=True)

df2
```
